```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('players.csv')
df.head()
```

```python
df_euw = pd.read_csv('players_euw.csv')
df_euw.head()
```

```python
import matplotlib.pyplot as plt

# Plotting boxplot for df_euw['lp']
plt.boxplot(df_euw['lp'], positions=[1], widths=0.6)
# Plotting boxplot for df['lp']
plt.boxplot(df['lp'], positions=[2], widths=0.6)

# Adding labels to the x-axis
plt.xticks([1, 2], ['EUW', 'EUNE'])

# Adding a title and labels to the axes
plt.title('LP distribution among top 500 players')
plt.xlabel('Dataframe')
plt.ylabel('lp')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Creating a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Plotting bar plot for df_euw['most_played_champ_1'] on the first subplot
df_euw['most_played_champ_1'].value_counts().head(20).plot.barh(ax=ax1 )
ax1.set_title('EUW top 20 champions among the top 500 players')
ax1.set_ylabel('Champion')
ax1.set_xlabel('Number of mains')

# Plotting bar plot for df['most_played_champ_1'] on the second subplot
df['most_played_champ_1'].value_counts().head(20).plot.barh(ax=ax2)
ax2.set_title('EUNE top 20 champions among the top 500 players')
ax2.set_ylabel('Champion')
ax2.set_xlabel('Number of mains')

# Adjusting the spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()
```

```python
games = pd.read_csv('games.csv')
games.head()
```

```python
games.columns
```

```python
games_first_euw  =pd.read_csv('games_first_euw.csv')
```

```python
first_item_cols =["blue team player 5 item 1" , "blue team player 4 item 1" , "blue team player 3 item 1" , "blue team player 2 item 1" , "blue team player 1 item 1" , "red team player 5 item 1" , "red team player 4 item 1" , "red team player 3 item 1" , "red team player 2 item 1" , "red team player 1 item 1"]

games_first_euw  =pd.read_csv('games_first_euw.csv')
all_items_euw = games_first_euw[first_item_cols].values.reshape(-1).tolist()
all_items_euw = pd.DataFrame(all_items_euw)


all_items = games[first_item_cols].values.reshape(-1).tolist()
all_items = pd.DataFrame(all_items)

# Creating a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Plotting bar plot for df_euw['most_played_champ_1'] on the first subplot
all_items_euw.value_counts().head(10).plot.barh(ax=ax1 )
ax1.set_title('EUW top 10 first item')
ax1.set_ylabel('Item')
ax1.set_xlabel('Number of games')

# Plotting bar plot for df['most_played_champ_1'] on the second subplot
all_items.value_counts().head(10).plot.barh(ax=ax2)
ax2.set_title('EUNE top 10 first item')
ax2.set_ylabel('Item')
ax2.set_xlabel('Number of games')

# Adjusting the spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()
```

```python
second_item_cols = ["blue team player 5 item 2", "blue team player 4 item 2", "blue team player 3 item 2", "blue team player 2 item 2", "blue team player 1 item 2", "red team player 5 item 2", "red team player 4 item 2", "red team player 3 item 2", "red team player 2 item 2", "red team player 1 item 2"]

games_first_euw = pd.read_csv('games_first_euw.csv')
all_items_euw = games_first_euw[second_item_cols].values.reshape(-1).tolist()
all_items_euw = pd.DataFrame(all_items_euw)

all_items = games[second_item_cols].values.reshape(-1).tolist()
all_items = pd.DataFrame(all_items)

# Creating a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Plotting bar plot for df_euw['most_played_champ_1'] on the first subplot
all_items_euw.value_counts().head(10).plot.barh(ax=ax1)
ax1.set_title('EUW top 10 second item')
ax1.set_ylabel('Item')
ax1.set_xlabel('Number of games')

# Plotting bar plot for df['most_played_champ_1'] on the second subplot
all_items.value_counts().head(10).plot.barh(ax=ax2)
ax2.set_title('EUNE top 10 second item')
ax2.set_ylabel('Item')
ax2.set_xlabel('Number of games')

# Adjusting the spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()
```

```python
first_rune_cols = ["blue team player 5 rune 1", "blue team player 4 rune 1", "blue team player 3 rune 1", "blue team player 2 rune 1", "blue team player 1 rune 1", "red team player 5 rune 1", "red team player 4 rune 1", "red team player 3 rune 1", "red team player 2 rune 1", "red team player 1 rune 1"]

games_first_euw = pd.read_csv('games_first_euw.csv')
all_runes_euw = games_first_euw[first_rune_cols].values.reshape(-1).tolist()
all_runes_euw = pd.DataFrame(all_runes_euw)

all_runes = games[first_rune_cols].values.reshape(-1).tolist()
all_runes = pd.DataFrame(all_runes)

# Creating a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Plotting bar plot for df_euw['most_played_champ_1'] on the first subplot
all_runes_euw.value_counts().head(10).plot.barh(ax=ax1)
ax1.set_title('EUW top 10 first rune')
ax1.set_ylabel('Rune')
ax1.set_xlabel('Number of games')

# Plotting bar plot for df['most_played_champ_1'] on the second subplot
all_runes.value_counts().head(10).plot.barh(ax=ax2)
ax2.set_title('EUNE top 10 first rune')
ax2.set_ylabel('Rune')
ax2.set_xlabel('Number of games')

# Adjusting the spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()
```
